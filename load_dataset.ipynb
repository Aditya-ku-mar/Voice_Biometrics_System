```python
import torchaudio

subsets = [
    "train-clean-100",
    "dev-clean",
    "test-clean"
]

for subset in subsets:
    print(f"Downloading {subset}...")
    torchaudio.datasets.LIBRISPEECH(
        root="data",
        url=subset,
        download=True
    )
    print(f"{subset} downloaded successfully!\n")

print("All datasets downloaded successfully.")
```
